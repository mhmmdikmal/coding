```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#Dataset untuk bar plot
tahun = np.array([2017, 2018, 2019, 2020,2021])
jumlah = np.array([121, 135, 321, 911, 1200])

#membuat figure dan axis
fig, ax = plt.subplots()

#membuat bar plot
ax.bar(tahun, jumlah)

#membuat judul dan label sumbu x dan y
ax.set(
    xlabel = 'tahun',
    ylabel = 'jumlah',
    title = 'jumlah penjualan e-book dari tahun 2017 s.d 2021 di online shop'
    )

#menampilkan bar plot
plt.show()
```
