```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Dataset untuk Scatter plot
dosen = np.array([10, 15, 20, 26, 17, 34, 40, 14])
mahasiswa = np.array([340,360,453,390,400, 425, 600,320])

#membuat figure dan axis
fig, ax = plt.subplots()

#membuat scatter plot
plt.scatter(dosen,mahasiswa)

#membuat judul
ax.set_title('Hubungan Antara Jumlah Dosen dengan Jumlah Mahasiswa di Harvard University')

plt.show()
```
